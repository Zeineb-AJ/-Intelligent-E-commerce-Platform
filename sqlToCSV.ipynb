```python
import csv
import re

# Chemin vers ton fichier .sql
sql_file_path = r'C:/Users/Zeineb/Desktop/TEKUP/ProjetBI/pro.sql'
csv_file_path = r'C:/Users/Zeineb/Desktop/TEKUP/ProjetBI/pro.csv'
```

```python

# Lire le fichier SQL
with open(sql_file_path, 'r', encoding='utf-8') as file:
    sql_content = file.read()

# Extraire la table cible et les colonnes
match = re.search(r'INSERT INTO `?(\w+)`? \((.*?)\) VALUES', sql_content)
if match:
    table_name = match.group(1)
    columns = [col.strip(' `') for col in match.group(2).split(',')]

# Extraire les valeurs
values = re.findall(r'VALUES\s*(\(.*?\));', sql_content, re.DOTALL)
if not values:
    values = re.findall(r'INSERT INTO .*? VALUES\s*(\(.*?\));', sql_content, re.DOTALL)

# Préparer les lignes
rows = []
for value_block in values:
    # Nettoyer et séparer les tuples
    tuples = re.findall(r'\((.*?)\)', value_block)
    for tup in tuples:
        fields = [f.strip(" '") for f in tup.split(',')]
        rows.append(fields)

# Sauvegarder en CSV
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(columns)  # en-têtes
    writer.writerows(rows)    # données

print(f"Fichier CSV généré : {csv_file_path}")
```
