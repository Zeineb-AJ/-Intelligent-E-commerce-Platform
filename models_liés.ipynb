```python
import numpy as np
import pandas as pd
import joblib
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
chemin_clf = 'C:/Users/Zeineb/Desktop/TEKUP/ProjetBI/modelF/modele_classificationFinallll.pkl'
chemin_scaler = 'C:/Users/Zeineb/Desktop/TEKUP/ProjetBI/modelF/scaler_classificationF.pkl'
chemin_reg = 'C:/Users/Zeineb/Desktop/TEKUP/ProjetBI/modelF/modele_discountFinalF.pkl'

clf_loaded = joblib.load(chemin_clf)
scaler_loaded = joblib.load(chemin_scaler)

reg = joblib.load(chemin_reg)
```

```python


def predire_etat_produit(stock, sales, age_en_jours, price):
    data_input = np.array([[stock, sales, age_en_jours, price]])
    data_scaled = scaler_loaded.transform(data_input)
    cluster_pred = clf_loaded.predict(data_scaled)[0]

    label_map = {
        0: "À liquider",
        1: "À liquider",
        2: "Normaux",
        3: "Performants"
    }

    return label_map.get(cluster_pred, "Autre")
```

```python
def predire_discount_si_a_liquider(stock, sales, age_en_jours, price):
    etat = predire_etat_produit(stock, sales, age_en_jours, price)

    if etat == "À liquider":
        data_input = np.array([[stock, sales, age_en_jours, price]])
        discount_pred = reg.predict(data_input)[0]
        return {
            "etat": etat,
            "discount": round(discount_pred * 100, 1)  # en pourcentage
        }
    else:
        return {
            "etat": etat,
            "discount": 0
        }
```

```python
result = predire_discount_si_a_liquider(stock=200, sales=15, age_en_jours=250, price=100)
print(result)
```

```python
produits = [
    (10, 5,50, 25),
    (400, 6, 100, 20),
    (600, 0, 500, 22)
]

for stock, sales, age, price in produits:
    res = predire_discount_si_a_liquider(stock, sales, age, price)
    if res['etat'].lower() == 'à liquider':
        print(f"Produit {res['etat'].upper()} - Stock: {stock}, Ventes: {sales}, Âge: {age} jours, Prix: {price} TND => Discount proposé: {res['discount']}%")
```
